### II.1. Définition et propriétés



Il faut imaginer une **file d’attente**, premier arrivé premier servi!
ceci se note **queue** ou **FIFO** first in first out in english

Il faut pouvoir :
* créer une file vide => init
* tester si une file est vide => isEmpty
* ajouter un élément à la file => enQueue
* lire le premier élément de la file sans la modifier => first
* retirer le premier élément de la file et le renvoyer => deQueue
    
**Chacune de ces opérations se fait en temps constant.**

- Il faudra ajouter les éléments dans la pile **instack**
- Il faudra lire ou défiler les éléments de la pile **outstack**.
- Lorsque l'on a besoin de **outstack** et que celle ci est vide, on retourne au préalable **instack** dans **outstack**.


```python
class Stack():
    def __init__(self):
        self.data = []
    
    def isEmpty(self):
        return self.data == []
    
    def push(self,element):
        self.data.append(element)
    
    def top(self):
        assert not self.isEmpty()
        return self.data[-1]
    
    def pop(self):
        assert not self.isEmpty()
        return self.data.pop()
```

```python
def init():
    instack = Stack()
    outstack = Stack()
    return (instack,outstack)

def isEmpty(queue):
    return queue[0].isEmpty() and queue[1].isEmpty()

def enQueue(element, queue):
    queue[0].push(element)
    
def first(queue):
    assert not isEmpty(queue)
    if queue[1].isEmpty():
        reverseInbox(queue)
    return queue[1].top()

def reverseInbox(queue):
    assert queue[1].isEmpty()
    while not queue[0].isEmpty():
        queue[1].push(queue[0].pop())

def deQueue(queue):
    assert not isEmpty(queue)
    if queue[1].isEmpty():
        reverseInbox(queue)
    return queue[1].pop()

# Un petit essai de fonctionnement:

q = init()
isEmpty(q)

enQueue(4,q)
enQueue(5,q)
isEmpty(q)

deQueue(q)

isEmpty(q)

deQueue(q)

isEmpty(q)
```
