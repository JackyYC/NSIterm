Faites la liste des nombres impairs divisible par 3 entre 10 et 30

```python
l = [i*2 + 1 for i in range (10//2,30//2) if (i*2 + 1) % 3 == 0]
l
```

```python
l = [i for i in range(10,30) if i % 2 == 1 and i % 3 == 0]
l
```

```python
l = [10 + i*2 +1 for i in range(10) if (10 + i*2 +1) % 3 == 0]
l
```

```python
l = [3 * i for i in range(10) if (3 * i) % 2 == 1 and 10 <= 3*i <= 30]
l
```
