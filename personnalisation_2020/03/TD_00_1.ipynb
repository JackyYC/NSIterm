```python
# Question 5
def tri_par_insertion(tableau):
    for k in range(1,len(tableau)):
        temp=tableau[k]
        j=k
        while j>0 and temp<tableau[j-1]:
            tableau[j]=tableau[j-1]
            j-=1
        tableau[j]=temp # C'est juste mais cela aurati nécessité quelques commentaires
    return tableau

# Et des tests de fonctionnement, par exemple celui de la video
tri_par_insertion([3,0,1,8,7,2,5,4,9,6])
```
