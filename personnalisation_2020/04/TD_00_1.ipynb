1. Ecrire une fonction **binaire** qui prend comme argument un int et qui renvoie une chaine de caract√®re repr√©sentant le binaire correspondant. Pour rappel, la conversion se fait par divisions successives par deux.
Par exemple: binaire(2) doit renvoyer "10" et binaire(7) "111". Votre fonction devra √™tre correctement comment√©e et document√©e.

```python
# Question 1.

def binaire(val): # Il faut utiliser les tabulations pour indenter
  binary = ""
  while val > 0:
    mod = val % 2
    val = val // 2
    if mod == 1:
      binary = "1" + binary
    else:
      binary = "0" + binary

  print(binary)
  return
```

```python
# Question 5

floats = [] # Il faut mettre des valeurs dans la liste si vous voulez tester votre fonction

def tri_insertion(liste):
    for i in range (1, len(liste)):
      y = i
      while liste[y] < liste[y - 1] and y != 0: # Il vaut mieux mettre y != 0 (ou mieux y > 0 qui sera plus lisible) en premier car python sortira sans tester la condition suivante qui n'a alors pas de sens
        liste[y], liste[y - 1] = liste [y - 1], liste [y]
        y -= 1
    print(liste)
    return()

tri_insertion(floats)
```
