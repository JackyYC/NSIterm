## Exercice 1. Compter les arbres binaires

Un arbre binaire est un arbre pour lequel chaque sommet possède 0 ou 2 fils. Un **noeud** est un sommet possédent deuxfils et une **feuille** est un sommet sans fils. La figure ci-dessous donne trois exemples d'abres binaires pour 0, 1 et 4 noeuds.
![arbres](img/arbres_binaires.png)

1. Ecrire une fonction tree qui prend comme argument un entier **n** représentant le nombre de noeuds et qui renvoie un entier correspondant au nombre d'arbres binaires possibles à n noeuds.
2. Dessinez au brouillon tous les arbres possibles pour 3 et 4 noeuds et vérifiez que votre fonction renvoie bien le bon résultat.
3. Déterminez tree(100) -il faudra peut être améliorer la fonction écrite dans un premier temps-


```python
# question 1.

def tree(n):
    assert type(n) == int
    assert n >= 0
    
    if n == 0:
        return 1
    
    else:
        total = 0
        for k in range (n): # k représnte le nombre de noeuds du coté droit du premier noeud.
            total += tree(n - 1 - k) * tree(k)
        
        return total
    
tree(3)
```

```python
# question 3.

memory = [1]
def tree(n):
    global memory
    assert type(n) == int
    assert n >= 0
    
    if n >= len(memory):
        total = 0
        for k in range(n):
            total += tree(n - 1 - k) * tree(k)
        memory.append(total) # Enregistre les calculs dans le tableau pour ne pas les refaire en boucle.
        
    return memory[n]

tree(100)
```

```python
def fib(n, a = 1, b = 1, k = 1):
    """L'argument a représente le résultat de la fonction fib avec n = k
    b est la valeur pour n = k - 1"""
    assert type(n) == int
    assert n >= 0
    
    if n == 0:
        return 1
    
    elif n == 1:
        return 1
    
    elif k == n:
        return a
    
    else:
        return fib(n, a+b, a, k+1) # K augmente et a prends la valeur de fib(k + 1)
    
fib(1000)
```

## Exercice 2. La fonction puissance

1. Ecrire une fonction **puissance** récursive qui prend en argument un float **x** et un entier positif ou nul **n** et qui renvoie la valeur de $x^n$ calculée par $x^0 = 1$ et $x^n = x \times x^{n-1}$

2. Ecrire une fonction **puissance_rapide** récursive qui prend en argument un float **x** et un entier positif ou nul **n** et qui renvoie la valeur de $x^n$ calculée par $x^0 = 1$ et $x^n = x^{n//2} \times x^{n//2}$ si n est pair et $x^n = x \times x^{n//2} \times x^{n//2}$ si n est impair.

3. Expliquer ce qu'apporte la deuxième fonction par rapport à la première.

```python
# Question 1.
def puissance(x, n):
    """
    x est un float
    n est un int
    Fonction récursive qui calule x**n et renvoie le resultat
    """
    
    assert type(x) == int or type(x) == float
    assert type(n) == int
    assert n >= 0
    
    if n == 0:
        return 1
    
    else:
        return x * puissance(x, n-1)
    
puissance(4, 3)
puissance(5, 2)
puissance(2, 4)
```

```python
# Question 2.
def puissance_rapide(x, n):
    """
    x est un float
    n est un int
    Fonction récursive qui calule x**n et renvoie le resultat
    """
    
    assert type(x) == int or type(x) == float
    assert type(n) == int
    assert n >= 0
    
    if n == 0:
        return 1
    
    else: # Plus rapide que la méthode aves le stockage de mémoire
        y = puissance_rapide(x, n//2) # Pour tout n € N, puissance_rapide(x, n) = x ** n
        return x ** (n%2) * y * y 
    
puissance_rapide(4, 3)
puissance_rapide(5, 2)
puissance_rapide(2, 4)
puissance_rapide(2, 0)
```

```python
def puissance_rapide_term(x, n, a = 1):
    """
    x est un float
    n est un int
    a sera le résultat mais ne pas assigné de valeur à a
    Fonction récursive qui calule x**n et renvoie le resultat
    """
    
    assert type(x) == int or type(x) == float
    assert type(n) == int
    assert n >= 0
    
    if n == 0:
        return a
    
    else:
        return puissance_rapide_term(x**2, n // 2, a * x**(n%2))
    
        # a est le résultat et se multiple par x ^0 ou ^1 dépendant de n%2
        # x devient son carré à chaque cran.
        
        # Donc on a pour n = 26 :
        # 1 * 1 * (x ** 2) * 1 * (x ** 8) * (x ** 16)
        
        # Donc on a pour n = 31 :
        # 1 * x * (x ** 2) * (x ** 4) * (x ** 8) * (x ** 16) 
        
        # Si n est pair on multiplie par 1, si n est impair on multiplie par x (au cran ou on est)
        
    
puissance_rapide_term(4, 3)
puissance_rapide_term(5, 2)
puissance_rapide_term(2, 4)
puissance_rapide_term(2, 0)
```

## Exercice 3. Recherche de caractère

1. Ecrire une fonction récursive **cherche** qui prend comme argument une chaine de caractères **ligne** et un caractère **cara** et qui renvoie le nombre d'occurences de la lettre dans la ligne.
2. Ecrire une fonction récursive **indice** qui prend comme argument une chaine de caractères **ligne** et un caractère **cara** et qui renvoie l'indice de la première occurence du caractère dans la ligne et None s'il n'y est pas.

```python
# question 1

def cherche(line, cara):
    
    """
    line est un string
    cara est un caractère simple
    Renvoie le nombre de caractères cara dans le string line
    """
        
    assert type(line) == str
    assert type(cara) == str
    assert len(cara) == 1
    
    if len(line) == 0 :
        return 0
    
    else:
        return (line[0] == cara) + cherche(line[1:], cara)
    
cherche("okay mek", "k")
```

```python
def cherche_term(line, cara, n = 0):
    
    """
    line est un string
    cara est un caractère simple
    Renvoie le nombre de caractères cara dans le string line
    """
    
    assert type(line) == str
    assert type(cara) == str
    assert len(cara) == 1
    
    if len(line) == 0 :
        return n
    
    else:
        return cherche(line[1:], cara, n + (line[0] == cara))
    
cherche_term("okay mek", "k")
```

La récursivité terminale permet de faire les calculs d'un coup au lieu de tour mettre en attente jusqu'à la fin de la pile.

```python
def indince_term(line, cara, index = 0):
    
    assert type(line) == str
    assert type(cara) == str
    assert len(cara) == 1
    
    if len(line) == 0:
        return None
    
    elif line[0] == cara:
        return index
    
    else:
        return indice_term(line[1:], cara, index + 1)
    
cherche_term("okay mek", "k")
```
