```python
def binaire(lettre):
    assert len(lettre) == 1
    assert 65 <= ord(lettre) <= 90 or ord("a") <= ord(lettre) <= ord("z")
    
    lettre = lettre.upper() 
    c = bin(ord(lettre))[2:]
    n = len(c)
    while n < 8 :
        c = "0" + c
        n += 1
        
    return c

binaire("a")
```

```python
def code(texte):
    assert type(texte) == str
    reponse = []
    for lettre in texte:
        reponse.append(binaire(lettre))
    return reponse

code("AABC")
```

```python
def taille(texte):
    assert type(texte) == str
    n = len(texte)
    return 8*n

taille("TDCOMPRESSIONDEDONNEESENNSI")
```

```python
def occurences(texte):
    assert type(texte) == str
    result = [0]*26
    for lettre in texte:
        result[indice(lettre)] += 1
        
    return result

occurences("AAABBC")
```

```python
def indice_mini(valeurs):
    m= min(valeurs)
    i = valeurs.index(m)
    return i

indice_mini([3, 2, 1, 2])
```
